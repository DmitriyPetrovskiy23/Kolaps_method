```python
width_rectangle = max_y - min_y
length_rectangle = max_x - min_x
print('Ширина прямоугольника = ', width_rectangle)
print('Длина прямоугольника = ', length_rectangle)

diagonal_rectangle = (width_rectangle ** 2 + length_rectangle ** 2) ** 0.5
print('Диагональ прямоугольника = ', diagonal_rectangle)
```
